### generate the datasets for MNIST addition with 2/3/4 images

```python
#load data
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5, ))])
dataset = torchvision.datasets.MNIST(root='./data/', train=True, download=True, transform=transform)
```

```python
print(dataset.data.shape)
print(dataset.targets.shape)


k = 3 #2,3,4
mode = 'train' #train/val/test


if mode == 'train':
    indices = np.random.permutation(60000) #train
    sums = np.zeros((int(60000/k),), dtype=np.int64)
else:
    indices = np.random.permutation(10000) #train
    sums = np.zeros((int(10000/k),), dtype=np.int64)


#uncomment to use

#for i in range(0,20000):

file_path = "data/labels/"+mode+"_data_s"+str(k)+".txt"
print(file_path)
f = open(file_path, "w")

dataset_size = 10000


for i in range(0,dataset_size):
    sum = 0
    for j in range(k):
        sum += dataset.targets[indices[i+j*dataset_size]]

    #print(indices[i], indices[i+dataset_size], indices[i+2*dataset_size])
    # sum = dataset.targets[indices[i]]+ dataset.targets[indices[i+dataset_size]]+ dataset.targets[indices[i+2*dataset_size]] + dataset.targets[indices[i+3*dataset_size]]+ dataset.targets[indices[i+4*dataset_size]] + dataset.targets[indices[i+5*dataset_size]]

    #sum = dataset.targets[indices[i]]+ dataset.targets[indices[i+dataset_size]]+ dataset.targets[indices[i+2*dataset_size]] + dataset.targets[indices[i+3*dataset_size]]+ dataset.targets[indices[i+4*dataset_size]]
    #sum = dataset.targets[indices[i]]+ dataset.targets[indices[i+dataset_size]]+ dataset.targets[indices[i+2*dataset_size]] + dataset.targets[indices[i+3*dataset_size]]
    #sum = dataset.targets[indices[i]]+ dataset.targets[indices[i+dataset_size]]+ dataset.targets[indices[i+2*dataset_size]] 

    #print(sum)
    line = ""
    for j in range(k):
        line += str(indices[i+j*dataset_size])+" "
    line += str(sum.numpy())
    # line = str(indices[i])+" "+ str(indices[i+]) + " "+str(indices[i+2*dataset_size])+" "+str(indices[i+3*dataset_size])+" "+str(indices[i+4*dataset_size])+" "+str(indices[i+5*dataset_size])+" "+ str(sum.numpy())
    #line = str(indices[i])+" "+ str(indices[i+dataset_size]) + " "+str(indices[i+2*dataset_size])+" "+str(indices[i+3*dataset_size])+" "+str(indices[i+4*dataset_size])+" "+ str(sum.numpy())
    #line = str(indices[i])+" "+ str(indices[i+dataset_size]) + " "+str(indices[i+2*dataset_size])+" "+str(indices[i+3*dataset_size])+" "+ str(sum.numpy())
    #line = str(indices[i])+" "+ str(indices[i+dataset_size]) + " "+str(indices[i+2*dataset_size])+" "+ str(sum.numpy())

    print(line)

    f.write(line)
    if i < dataset_size -1: 
        f.write("\n")

f.close()

```
